```python
# target variable encoding
df["make"] = df.groupby("make")["selling_price"].transform("mean") 
df.make
# calculates the mean of the selling price of the make group
# groupby command is used to group similar kind of entities
# pandas in sql in python
# transform is the commmand which will convert the categorical variable into a continous one
```

```python
df["model"] = df.groupby("model")["selling_price"].transform("mean")
df.model
```

```python
df
```

```python
# normalisation(scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1 = pd.DataFrame(scaler.fit_transform(df), columns= df.columns)
df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df1["selling_price"]
x = df1.drop("selling_price",axis = 1)
y.shape, x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train.shape, y_train.shape, x_test.shape, y_test.shape # test data to predict the output
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # linearregression is the class and model is the object of the class
model.fit(x_train, y_train) # call method fit and it will train the model using the trained data
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# to predict the price of car
y_test_predict = model.predict(x_test) #command to predict the test data
y_test_predict 
```

```python
model.score(x_test, y_test) # to find the accuracy of the model
```

```python
fig = plt.figure()
plt.scatter(y_test_predict, y_test) # graph between predicted and the test output
plt.legend()
plt.show()
```

```python
import seaborn as sns

imp = pd.DataFrame(list(zip(x_test.columns, np.abs(model.coef_))),
                   columns=['feature', 'coeff'])
sns.barplot(x = 'feature', y = 'coeff', data = imp)
plt.xticks(rotation = 80)
```
